```python
df=pd.read_csv("sales_master_processed.csv")
```

```python
#Top 10 Customers
df.groupby(['CustomerKey','FullName'])['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
#Revenue by Age Group
df.groupby('Agegroup')['Sales'].sum()
```

Insight

The 56–65 age segment is the most valuable customer group.
Nearly 40% of total revenue comes from this age bracket.

Business Recommendation
1.Focus marketing campaigns on customers aged 56–65.
2.Investigate which products this segment purchases most.
3.Create loyalty programs tailored to this demographic.

```python
#Revenue by Income Group
df.groupby('IncomeGroup')['Sales'].sum()
```

This suggests:

The company's products appeal to middle-income consumers.
The customer base may contain many more middle-income customers.
High-income customers may represent a smaller segment.

Key Findings

Revenue reached $24.9M across 25,164 orders.
Bikes contributed ~95% of revenue and profit.
Professionals were the highest-value customer segment.
Customers aged 56–65 generated the largest share of revenue.
US and Australia were the strongest markets.
Mountain-200 products were the top-performing product line.
Return-rate analysis identified several bike models and apparel products requiring further investigation.

Business Recommendations

Increase accessory and clothing cross-selling.
Focus marketing on Professionals and Management occupations.
Replicate successful US and Australia strategies in lower-performing markets.
Investigate high-return-rate bike and apparel products.
Develop retention programs for high-value customer segments.
